```python
# Scrape each article as a unit
import requests
from bs4 import BeautifulSoup

url = "https://u.today/search/node?keys=bitcoin"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

articles = soup.find_all("div", class_="news__item")

titles = []
dates = []

for article in articles:
    title_tag = article.find("div", class_="news__item-title")
    date_tag = article.find("div", class_="humble")

    if title_tag and date_tag:
        titles.append(title_tag.get_text(strip=True))
        dates.append(date_tag.get_text(strip=True))

# Output
for title, date in zip(titles, dates):
    print(f"{date} - {title}")
```

```python
# Scrape each article as a unit and include author and link
import requests
from bs4 import BeautifulSoup

url = "https://u.today/search/node?keys=bitcoin"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

articles = soup.find_all("div", class_="news__item")

titles = []
dates = []
authors = []
links = []

for article in articles:
    title_tag = article.find("div", class_="news__item-title")
    date_tag = article.find("div", class_="humble")
    author_tag = article.find("div", class_="humble humble--author")
    link_tag = title_tag.find("a") if title_tag else None

    if title_tag and link_tag:
        titles.append(title_tag.get_text(strip=True))
        dates.append(date_tag.get_text(strip=True))
        authors.append(author_tag.get_text(strip=True))
        links.append("https://u.today" + link_tag['href'])

# Output
for title, date, author, link in zip(titles, dates, authors, links):
    print(f"{date} - {title} by {author} - Link: {link}")
```

```python
# Get the titles, dates, authors, and links of articles from u.today

import pandas as pd  # Add this import at the top

# BEFORE your for loop:
dates = []
titles = []
authors = []
links = []

for article in articles:
    link_tag = article.find("a", class_="news__item-body")
    title = link_tag.get_text(strip=True) if link_tag else "N/A"
    link = link_tag["href"] if link_tag else "N/A"

    author_tag = article.find("a", class_="humble humble--author")
    author = author_tag.get_text(strip=True) if author_tag else "N/A"

    date_tags = article.find_all("div", class_="humble")
    date = date_tags[-1].get_text(strip=True) if date_tags else "N/A"

    # Append each scraped value to the respective list
    dates.append(date)
    titles.append(title)
    authors.append(author)
    links.append(link)

    print(f"{date} - {title} by {author} - Link: {link}")

# Create a DataFrame to store the results
df = pd.DataFrame({
    "date": dates,
    "title": titles,
    "author": authors,
    "link": links
})

print(df)  
```

## Adding missing articles manually to the database

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import sqlite3

# --- Helper function to parse full date with time ---
def parse_date(date_str):
    try:
        return datetime.strptime(date_str, "%b %d, %Y - %H:%M")
    except Exception:
        return None

# --- Scrape the page ---
url = "https://u.today/search/node?keys=bitcoin"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
articles = soup.find_all("div", class_="news__item")

# --- Prepare lists ---
new_dates = []
new_titles = []
new_authors = []
new_links = []

for article in articles:
    link_tag = article.find("a", class_="news__item-body")
    title = link_tag.get_text(strip=True) if link_tag else "N/A"
    link = "https://u.today" + link_tag["href"] if link_tag else "N/A"

    author_tag = article.find("a", class_="humble humble--author")
    author = author_tag.get_text(strip=True) if author_tag else "N/A"

    date_tags = article.find_all("div", class_="humble")
    date_str = date_tags[-1].get_text(strip=True) if date_tags else "N/A"
    parsed_date = parse_date(date_str)

    # Filter: June 9 to June 12, 2025 (inclusive)
    if parsed_date and datetime(2025, 6, 9) <= parsed_date <= datetime(2025, 6, 12):
        new_dates.append(parsed_date)
        new_titles.append(title)
        new_authors.append(author)
        new_links.append(link)

# --- Create DataFrame ---
new_df = pd.DataFrame({
    "date": new_dates,
    "title": new_titles,
    "author": new_authors,
    "link": new_links
})

# --- Remove already stored articles from SQLite ---
conn = sqlite3.connect('articles.db')
try:
    existing_df = pd.read_sql_query("SELECT link FROM articles", conn)
    existing_links = set(existing_df["link"])
except Exception:
    existing_links = set()

new_df = new_df[~new_df["link"].isin(existing_links)]

# --- Insert new rows ---
if not new_df.empty:
    new_df.to_sql("articles", conn, if_exists="append", index=False)
    print(f"✅ Added {len(new_df)} new articles to the database.")
else:
    print("⚠️ No new articles found for the specified date range.")

conn.close()
```

```python
import sqlite3
from datetime import datetime

# Connect to your local SQLite DB
conn = sqlite3.connect('articles.db')
cur = conn.cursor()

# Select all rows with id and date
cur.execute("SELECT rowid, date FROM articles")
rows = cur.fetchall()

for rowid, date_str in rows:
    try:
        # Parse the ISO date string (or try other formats if needed)
        parsed_date = datetime.fromisoformat(date_str)
        # Format to desired string format
        formatted_date = parsed_date.strftime("%b %d, %Y - %H:%M")

        # Update the row with formatted date
        cur.execute("UPDATE articles SET date = ? WHERE rowid = ?", (formatted_date, rowid))
        print(f"Updated row {rowid} date to {formatted_date}")

    except Exception as e:
        print(f"Failed to parse/update row {rowid} date '{date_str}': {e}")

conn.commit()
cur.close()
conn.close()
```

```python
import sqlite3
from datetime import datetime

def parse_custom_date(date_str):
    try:
        # Try your current format first
        return datetime.strptime(date_str, "%b %d, %Y - %H:%M")
    except Exception:
        # Try ISO if it is already there
        try:
            return datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
        except Exception:
            return None

conn = sqlite3.connect('articles.db')
cur = conn.cursor()

cur.execute("SELECT rowid, date FROM articles")
rows = cur.fetchall()

for rowid, date_str in rows:
    parsed_date = parse_custom_date(date_str)
    if parsed_date:
        # Convert to Supabase format
        fixed_date = parsed_date.strftime("%Y-%m-%d %H:%M:%S")
        cur.execute("UPDATE articles SET date = ? WHERE rowid = ?", (fixed_date, rowid))
        print(f"Row {rowid}: '{date_str}' -> '{fixed_date}'")
    else:
        print(f"Row {rowid}: Could not parse date '{date_str}'")

conn.commit()
cur.close()
conn.close()
```
